## Pre Processing Data

### Drop Director and Cast

Because I don't need the director and cast column so I could just drop it since the majority of the null value came from director and cast table

```python
netflix.drop(['director', 'cast'], axis = 1, inplace = True)
```

```python
netflix.isnull().sum()
```

### Fill the null Value in Country

first we need to find a modus from country.

the modus will be used to fill null value in country

```python
netflix['country'].mode()
```

modus from country data is united states, so we fill null value with united states

```python
netflix['country'] = netflix['country'].fillna(netflix['country'].mode()[0])
```

After we fill null value, check is there is still null value in country

```python
netflix.isnull().sum()
```

### Rating

find null value in rating

```python
null_rating = netflix['rating'].isnull()
null_rating
```

collect all null value in new dataframe

```python
null_ratingDf = netflix[null_rating]
null_ratingDf
```

because it is only 7 null value in rating, so we'll replace it with real rating data about the movie.

```python
replace_rating = {
    67 : 'TV-PG', # id, rating
    2359 : 'TV-14',
    3660 : 'TV-MA',
    3736 : 'TV-MA',
    3737 : 'NR',
    3738 : 'TV-MA',
    4323 : 'TV-MA'
}
```

```python
replace_rating.items()
```

make a loop to find id and rating in netflix dataframe and replace it with replace_rating.items() that we made before

```python
for id, rate in replace_rating.items():
  netflix.loc[id, 'rating'] = rate
```

```python
netflix.isnull().sum()
```

### Date Added

Replace all null/nan value in date_added with Not Added

```python
netflix['date_added'] = netflix[['date_added']].replace(np.nan, 'Not Added')
```

```python
netflix.isnull().sum()
```

### Release month

because there is no release month in netflix dataframe, I added a new column named month_released.

To find the release month of the movie, i use date_added to find the month. I use lambda to split the first string of data_added.

lambda x : x.lstrip().split(" ")[0] used to find string[0] or the first string and stop when it met " " or space.

```python
netflix['month_released'] = netflix['date_added'].apply(lambda x : x.lstrip().split(" ")[0])
netflix.head()
```

```python
plt.figure(figsize = (14, 8))
sns.countplot(data = netflix, x = 'month_released', order = netflix['month_released'].value_counts().index[0:12], palette = 'crest', hue = 'month_released')
plt.title('Month Released', fontsize = 12, fontweight = 'bold')
plt.xlabel('Month')
plt.xticks(rotation = 45)
plt.ylabel('Total')
```
